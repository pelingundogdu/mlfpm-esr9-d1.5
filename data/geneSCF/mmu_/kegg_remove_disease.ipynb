```python
import pandas as pd
```

```python
# df_mmu = pd.read_csv('kegg_database_mmu.txt', error_bad_lines=False, sep='~|\t', header=None)
# print(len(df_mmu))

df = pd.read_csv('kegg_database.txt', error_bad_lines=False, sep='~|\t', header=None)
print(len(df))
df.columns = ['pathway_id', 'pathway_name', 'pathay_genes']
```

```python
list_genes = [df.iloc[i]['pathay_genes'] for i in range(len(df))]

flat_list = [val for sublist in list_genes for val in sublist.split(',')]
del(list_genes)
flat_list.remove('')
```

```python
len(set(flat_list))
```

```python
sorted(df.iloc[:, 1].tolist())
```

```python
df['cancer'] = df.iloc[:, 1].str.find('cancer')
df['disease'] = df.iloc[:, 1].str.find('disease')
df['kept'] = 0
df.loc[(df['cancer']==-1) & (df['disease']==-1), 'kept'] = 1
df.groupby('kept').size()
```

```python
df[['pathway_id','pathway_name', 'kept']].to_csv('kegg_database_wo_disease_related.csv')
```

```python
df_edit = pd.read_csv('kegg_database_wo_disease_related_Inma_edit.csv', index_col='Unnamed: 0')
df_edit['edit'].fillna(1, inplace=True)
df_edit.loc[df_edit['kept'] == 0, 'edit'] = 0
df_edit
```
