```python
# Details about this code   ;   - exporting all genes information from geneSCF KEGG database 
# Exported folder location ;  
#   1.  ../data/pathways        > stores genes list (csv format)
```

```python
import pandas as pd
import warnings
import os
warnings.filterwarnings('ignore')
```

```python
path_pathways = '../data/pathways/'
```

```python
df_kegg = pd.read_csv(os.path.join(path_pathways + 'kegg_pathways_final.csv'), error_bad_lines=False, sep=',')#.drop(columns='Unnamed: 0')
df_kegg
```

```python
list_genes = [df_kegg.iloc[i]['pathway_genes'] for i in range(len(df_kegg))]
flat_list = [val for sublist in list_genes for val in sublist.split(',')]
del(list_genes)
flat_list.remove('')
list_unique = list(set(flat_list))
list_unique.remove('')

pd.DataFrame(list_unique).to_csv(os.path.join(path_pathways + 'kegg_gene_list.csv'), index=False)
```
